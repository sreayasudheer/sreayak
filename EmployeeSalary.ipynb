```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import joblib

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# Apply SMOTE on training data
smote = SMOTE(random_state=42)
x_train_bal, y_train_bal = smote.fit_resample(x_train, y_train)

# Define models with tuned parameters
models = {
    "LogisticRegression": LogisticRegression(max_iter=2000, solver='liblinear'),
    "RandomForest": RandomForestClassifier(n_estimators=150, max_depth=15, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=7),
    "SVM": SVC(kernel='rbf', C=1.5, probability=True),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=150, learning_rate=0.1, max_depth=4)
}

results = {}
best_model = None
best_score = 0

# Training and evaluation
for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(x_train_bal, y_train_bal)
    y_pred = pipe.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label='>50K')

    results[name] = {"accuracy": acc, "f1_score": f1}

    print(f"\n{name} Results")
    print(f"Accuracy: {acc:.4f}")
    print(f"F1-score: {f1:.4f}")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    # Save best model
    if acc > best_score:
        best_score = acc
        best_model = pipe
        best_model_name = name

# Save the best model for prediction
joblib.dump(best_model, "best_model.pkl")
print(f"\nâœ… Best model saved: {best_model_name} with Accuracy: {best_score:.4f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
import joblib

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Apply SMOTE to balance the training data
smote = SMOTE(random_state=42)
x_train_bal, y_train_bal = smote.fit_resample(x_train, y_train)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

# Train and evaluate
for name, model in models.items():
    model.fit(x_train_bal, y_train_bal)  # <- use balanced data
    preds = model.predict(x_test)        # <- test on original test set
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Get Best Model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\nBest model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("Saved best model as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np

# Load the trained model
model = joblib.load("best_model.pkl")

st.set_page_config(
    page_title="Employee Salary Classification",
    page_icon="ðŸ’¼",
    layout="centered"
)

st.title("ðŸ’¼ Employee Salary Classification App")
st.markdown("Predict whether an employee earns >50K or â‰¤50K based on input features")

# Define categorical mappings with consistent names
CATEGORICAL_MAPPINGS = {
    'workclass': ['Private', 'Self-emp-not-inc', 'Local-gov', 'Others', 'State-gov', 'Self-emp-inc', 'Federal-gov'],
    'marital-status': ['Married-civ-spouse', 'Never-married', 'Divorced', 'Separated', 'Widowed', 'Married-spouse-absent', 'Married-AF-spouse'],
    'occupation': ['Prof-specialty', 'Craft-repair', 'Exec-managerial', 'Adm-clerical', 'Sales', 'Other-service', 'Machine-op-inspct', 'Others', 'Transport-moving', 'Handlers-cleaners', 'Tech-support', 'Farming-fishing', 'Protective-serv'],
    'relationship': ['Husband', 'Not-in-family', 'Own-child', 'Unmarried', 'Wife', 'Other-relative'],
    'race': ['White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'],
    'gender': ['Male', 'Female'],  # Changed from 'sex' to 'gender'
    'native-country': [
        'United-States', 'Mexico', 'Others', 'Philippines', 'Germany', 'Canada', 'Puerto-Rico', 'El-Salvador',
        'India', 'Cuba', 'England', 'Jamaica', 'South', 'China', 'Italy', 'Dominican-Republic', 'Vietnam',
        'Guatemala', 'Japan', 'Poland', 'Columbia', 'Taiwan', 'Haiti', 'Iran', 'Portugal', 'Nicaragua',
        'Peru', 'France', 'Greece', 'Ecuador', 'Ireland', 'Hong', 'Trinadad&Tobago', 'Cambodia', 'Thailand',
        'Laos', 'Yugoslavia'
    ]
}

# Column order expected by the model
COLUMN_ORDER = [
    'age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status',
    'occupation', 'relationship', 'race', 'gender',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country'
]

# Manual label encoding function
def manual_label_encode(value, categories):
    try:
        return categories.index(value)
    except ValueError:
        return -1

# --- User Input ---
st.subheader("ðŸ‘¤ Personal & Work Information")

col1, col2 = st.columns(2)
with col1:
    age = st.slider("Age", 17, 90, 30)
    fnlwgt = st.number_input("Final Weight", 10000, 1500000, 200000)
    educational_num = st.slider("Educational Number", 1, 16, 9)  # Corrected name
    capital_gain = st.number_input("Capital Gain", 0, 100000, 0)
    capital_loss = st.number_input("Capital Loss", 0, 5000, 0)

with col2:
    hours_per_week = st.slider("Hours per Week", 1, 99, 40)
    workclass = st.selectbox("Work Class", CATEGORICAL_MAPPINGS['workclass'])
    marital_status = st.selectbox("Marital Status", CATEGORICAL_MAPPINGS['marital-status'])
    occupation = st.selectbox("Occupation", CATEGORICAL_MAPPINGS['occupation'])

st.subheader("ðŸ  Family & Demographic Info")
col3, col4 = st.columns(2)
with col3:
    relationship = st.selectbox("Relationship", CATEGORICAL_MAPPINGS['relationship'])
    race = st.selectbox("Race", CATEGORICAL_MAPPINGS['race'])

with col4:
    gender = st.radio("Gender", CATEGORICAL_MAPPINGS['gender'])  # Changed to 'gender'
    native_country = st.selectbox("Native Country", CATEGORICAL_MAPPINGS['native-country'])

# Predict button
if st.button("ðŸ” Predict Salary Class"):
    input_data = {
        'age': age,
        'workclass': manual_label_encode(workclass, CATEGORICAL_MAPPINGS['workclass']),
        'fnlwgt': fnlwgt,
        'educational-num': educational_num,  # Corrected to match model
        'marital-status': manual_label_encode(marital_status, CATEGORICAL_MAPPINGS['marital-status']),
        'occupation': manual_label_encode(occupation, CATEGORICAL_MAPPINGS['occupation']),
        'relationship': manual_label_encode(relationship, CATEGORICAL_MAPPINGS['relationship']),
        'race': manual_label_encode(race, CATEGORICAL_MAPPINGS['race']),
        'gender': manual_label_encode(gender, CATEGORICAL_MAPPINGS['gender']),  # Changed to 'gender'
        'capital-gain': capital_gain,
        'capital-loss': capital_loss,
        'hours-per-week': hours_per_week,
        'native-country': manual_label_encode(native_country, CATEGORICAL_MAPPINGS['native-country']),
    }

    # Create DataFrame with correct column order
    input_df = pd.DataFrame([input_data], columns=COLUMN_ORDER)

    try:
        prediction = model.predict(input_df)
        prediction_proba = model.predict_proba(input_df)

        st.success(f"Predicted Salary Class: {'>50K' if prediction[0] == 1 else '<=50K'}")
        st.info(f"Confidence: {max(prediction_proba[0]):.1%}")

    except Exception as e:
        st.error(f"Prediction Error: {str(e)}")
        st.warning("Check for missing or mismatched features in the input.")

# --- Batch Prediction ---
st.markdown("---")
st.header("ðŸ“‚ Batch Prediction")
st.info("CSV must contain these columns with exact names: " + ", ".join(COLUMN_ORDER))
uploaded_file = st.file_uploader("Upload a CSV file with employee data", type=["csv"])

if uploaded_file is not None:
    try:
        batch_data = pd.read_csv(uploaded_file)

        # Check for required columns
        missing_cols = set(COLUMN_ORDER) - set(batch_data.columns)
        if missing_cols:
            st.error(f"Missing columns in CSV: {', '.join(missing_cols)}")
            st.stop()

        # Encode categorical columns
        for col in CATEGORICAL_MAPPINGS:
            if col in batch_data.columns:
                batch_data[col] = batch_data[col].apply(
                    lambda x: manual_label_encode(x, CATEGORICAL_MAPPINGS[col])
                )

        # Reorder columns to match training data
        batch_data = batch_data[COLUMN_ORDER]

        st.subheader("Uploaded Data (Encoded Preview)")
        st.write(batch_data.head())

        if st.button("ðŸ“Š Run Batch Prediction"):
            with st.spinner("Processing..."):
                predictions = model.predict(batch_data)
                prediction_probas = model.predict_proba(batch_data)

                batch_data["PredictedClass"] = predictions
                batch_data["PredictedClass"] = batch_data["PredictedClass"].map({0: "<=50K", 1: ">50K"})
                batch_data["PredictionConfidence"] = [max(proba) for proba in prediction_probas]

                st.subheader("Prediction Results")
                st.write(batch_data.head())

                csv = batch_data.to_csv(index=False).encode('utf-8')
                st.download_button(
                    label="â¬‡ Download Predictions CSV",
                    data=csv,
                    file_name='salary_predictions.csv',
                    mime='text/csv'
                )
    except Exception as e:
        st.error(f"Batch Prediction Error: {str(e)}")

# Footer
st.markdown("---")
st.caption("This tool uses demographic and employment data to predict salary classification.")
```
